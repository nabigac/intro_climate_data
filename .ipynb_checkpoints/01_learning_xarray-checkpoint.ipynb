## building a basic n-d DataArray

```python
import pooch
url = "https://www.ldeo.columbia.edu/~rpa/float_data_4901412.zip"
files = pooch.retrieve(url, processor=pooch.Unzip(), known_hash="2a703c720302c682f1662181d329c9f22f9f10e1539dc2d6082160a469165009")
files
```

```python
date = np.load(files[0])
T = np.load(files[1])
S = np.load(files[2])
P = np.load(files[3])
levels = np.load(files[4])
lon = np.load(files[5])
lat = np.load(files[6])
```

```python
# organize data and coordinates of the salinity variable

da_salinity = xr.DataArray(S, dims=['level', 'date'],
                           coords={'level': levels,
                                   'date': date},)
da_salinity
```

```python
da_salinity.attrs['units'] = 'PSU'
da_salinity.attrs['standard_name'] = 'sea_water_salinity'
da_salinity
```

```python
da_salinity.plot(yincrease=False)
```
