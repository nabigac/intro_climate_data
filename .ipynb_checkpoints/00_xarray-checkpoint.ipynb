## building a basic n-d DataArray

```python
import pooch
url = "https://www.ldeo.columbia.edu/~rpa/float_data_4901412.zip"
files = pooch.retrieve(url, processor=pooch.Unzip(), known_hash="2a703c720302c682f1662181d329c9f22f9f10e1539dc2d6082160a469165009")
files
```

```python
date = np.load(files[0])
T = np.load(files[1])
S = np.load(files[2])
P = np.load(files[3])
levels = np.load(files[4])
lon = np.load(files[5])
lat = np.load(files[6])
```

```python
# organize data and coordinates of the salinity variable

da_salinity = xr.DataArray(S, dims=['level', 'date'],
                           coords={'level': levels,
                                   'date': date},)
da_salinity
```

```python
da_salinity.attrs['units'] = 'PSU'
da_salinity.attrs['standard_name'] = 'sea_water_salinity'
da_salinity
```

```python
da_salinity.plot(yincrease=False)
```

## building a Dataset

```python
ds = xr.Dataset(
    data_vars={
        'salinity':    (('level', 'date'), S),
        'temperature': (('level', 'date'), T),
        'pressure':    (('level', 'date'), P)
    },
    coords={
        'level': levels,
        'date': date
    }
)

ds
```

```python
ds.coords['lon'] = ('date', lon)
ds.coords['lat'] = ('date', lat)

ds = ds.set_coords(['lon', 'lat'])
```

```python
ds.salinity.sel(level=2).plot()
```

```python
ds.salinity.date
```

```python
ds.salinity.sel(date=slice('2013-10-01', '2013-12-01')).plot(yincrease=False)
```

```python
temp_kelvin = ds.temperature + 273.15
temp_kelvin.plot(yincrease=False, cmap='jet')
```

```python
g = 9.8
buoyancy = g * (2e-4 * ds.temperature - 7e-4 * ds.salinity)
buoyancy.plot(yincrease=False)
```

```python
ds.temperature.mean(axis=0)
```

```python
ds.mean(dim='date').salinity.plot(y='level', yincrease=False)
```

## more advanced xarray functions

```python
url = 'http://www.esrl.noaa.gov/psd/thredds/dodsC/Datasets/noaa.ersst.v5/sst.mnmean.nc'
ds = xr.open_dataset(url, drop_variables=['time_bnds'])
ds = ds.sel(time=slice('1960', '2018')).load()
ds
```

```python
ds["sst"].attrs
```

```python
ds.sst[-1].plot(vmin=-2, vmax=30)
```

```python
ds.time
```

```python
ds.sst.sel(lon=300, lat=50).plot()
```

```python
gb = ds.sst.groupby('time.month')
```

```python
sst_month_mean = gb.mean(dim='time')
```

```python
# climatology at a specific point in north atlantic

sst_month_mean.sel(lon=300, lat=50).plot()
```

```python
# zonal mean climatology

sst_month_mean.mean(dim='lon').transpose().plot.contourf(levels=12, vmin=-2, vmax=30, cmap="jet")
```

```python
# plot difference between december and july climatology

(sst_month_mean.sel(month=12) - sst_month_mean.sel(month=7)).plot(vmax=10)
```

```python
# tranformation: look at the anomaly

gb = ds.groupby('time.month')
ds_anomaly = gb - gb.mean(dim='time')
ds_anomaly
```

```python
ds_anomaly.sst.sel(lon=300, lat=50).plot()
```

```python
# transformation: calculate a 5 year mean using resample

ds_anomaly_resample = ds_anomaly.resample(time='5Y').mean(dim='time')
ds_anomaly_resample
```

```python
ds_anomaly.sst.sel(lon=300, lat=50).plot()
ds_anomaly_resample.sst.sel(lon=300, lat=50).plot(marker='o')
```

```python
# transformation: calculate a 12 month rolling average

ds_anomaly_rolling = ds_anomaly.rolling(time=12, center=True).mean()
ds_anomaly_rolling
```

```python
ds_anomaly.sst.sel(lon=300, lat=50).plot()
ds_anomaly_rolling.sst.sel(lon=300, lat=50).plot(color='k')
```
